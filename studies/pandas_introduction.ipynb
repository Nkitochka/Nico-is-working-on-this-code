```python
#__PANDAS__#
import pandas as pd
import numpy as np

nums_place_holder = [1, 2, 3, 4, 5, 6]
n = pd.Series(nums_place_holder) #default index (means that each item will be numberd from 0 in first column)
print(n, '\n')

n = pd.Series(nums_place_holder, index = [1, 2, 3, 4, 5, 6]) #custom index
print(n)
```

```python
# Dictionary to Series
dct = {'name':'Asabeneh',
       'country':'Finland',
       'city':'Helsinki'}
d = pd.Series(dct)
print(d)
```

```python
# const Series display
x = pd.Series(10, index = [1, 3, 5, 7])
print(x, '\n')

y = pd.Series("BONK!", index = [2, 4, 6, 8])
print(y)
```

```python
# __DataFrame__ #
users_data = [
    ['FishyNick345', 'PC', 34],
    ['KnocKnock!', 'Mobile', 102],
    ['FlushHana', 'Xbox', 611],
    ['Mibuuu51', 'PC', 433]
]

ud = pd.DataFrame(users_data, columns = ['Username', 'Device', 'Hours in-game'])
print(ud)
```
