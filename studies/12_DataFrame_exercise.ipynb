```python
# First exercise
df = pd.DataFrame({
    "name" : ['Anna', 'Ivan', 'Olga', 'Petr'],
    "age" : [23, 30, 27, 35],
    "city" : ["Moscow", "Kazan", "Moscow", "Samara"]
})
print(df, "\n\n")

print("First task:\n", df.head(3), '\n\n') # 1

print("Second task:\n", df[ df["age"] > 25], '\n\n') # 2

print("Thrird task:\n Средний возраст опрошенных равен ", df["age"].mean())
```

```python
# Second exercise
print(df[["name", "age"]], '\n\n') # 1

#print(df["age"].sort_values(), '\n\n') # 2 - not quite what was needed, but okay
print(df.sort_values("age"), '\n\n')

print(f'Max age is {df["age"].max()}. And the minimum age is {df["age"].min()}.')
```

```python
# Third exercise
print(df[ df['city'] == 'Moscow']) # 1

mean_age = df['age'].mean()
print(mean_age)
#print( df[ (df['city'] == 'Moscow') & ( df['age'] > df['age'].mean() ) ] ) - prototype
print( df[ (df['city'] == 'Moscow') & ( df['age'] > mean_age ) ] )
```
